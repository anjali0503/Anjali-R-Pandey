```
from keras.datasets import fashion_mnist #fashion_mnist is one of the dataset 
data = fashion_mnist.load_data() #loading a data.
```

```
(X_train,y_train), (X_test, y_test)=fashion_mnist.load_data() #now will divide dataset in testing and training .
```

**Keras** is a neural network library while **TensorFlowis** the open-source library for a number of various tasks in machine learning. TensorFlow provides both high-level and low-level APIs while Keras provides only high-level APIs. ... Both frameworks thus provide high-level APIs for building and training models with ease.

```
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import  Input, Dense
import numpy as np
```

import Sequential #when there is one input for one output
import dense #Dense is a deep neural network

```
x=np.array([[0,0], [0,1],[1,0], [1,1]],"uint8") 
y=np.array([[0],[0],[0],[1]],"uint8")
```

**uint8** is an unsigned 8-bit integer that can represent values 0.. 255. int on the other hand is usually a 32-bit signed integer

```
#specific to keras only, will not work in different dataset.
model=Sequential()
model.add(Dense(units=1,activation="sigmoid",input_dim=2)) #in this  units means number of neurons in each layer and input_dim is used for dimensions specify
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
#optimizer is an  argument which is used for keras  models.
#adam is an optimization algorithm in which models are handled properly.
#loss will show loss accuracy .
#binary_crossentropy' this is used for preventing overflowing or underflowing of numbers
```

```
model.fit(x,y,epochs=1000) #this is for training  model
```

```
X_train=X_train.reshape((X_train.shape[0],28*28)).astype('float32')
X_test=X_test.reshape((X_test.shape[0], 28*28)).astype('float32')
```

```
X_train=X_train/255
X_test=X_test/255
```

```
from keras.utils import np_utils #keras.utils se generators callbacks ye sab import hote
```

```
y_train=np_utils.to_categorical(y_train)
y_test=np_utils.to_categorical(y_test)
num_classes=y_test.shape[1]
num_classes
```

```
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```
model=Sequential()
model.add(Dense(32,input_dim=28*28,activation='relu')) #it is of model type of sequential in which  data stackwise gets arranged .
model.add(Dense(64,activation='relu'))#64 is to  change , layers get  change ,keep on doubling,64 double 128
model.add(Dense(128,activation='relu'))#this is the 3rd layer
model.add(Dense(256,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```
model.compile(loss="categorical_crossentropy",optimizer='adam',metrics=['accuracy']) #compiling mode
```

```
model.summary()
```

```
model.fit(X_train,y_train,epochs=10,batch_size=200)
```

```
scores=model.evaluate(X_test,y_test)
scores
```
